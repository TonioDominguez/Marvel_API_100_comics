# Obtaining columns

### Obtaining title

```python
comic_list = []
```

```python
for comic in com:
    comic_list.append(comic["title"])
```

```python
len(comic_list)
```

### Obtaining description

```python
comic_description = []
```

```python
for comic in com:
    comic_description.append(comic["description"])
```

```python
len(comic_description)
```

### Obtaining characters

```python
comic_characters = []
```

```python
for comic in com:
    comic_characters.append(', '.join([character["name"] for character in comic["characters"]["items"]]))
```

```python
len(comic_characters)
```

### Obtaining id

```python
comic_id = []
```

```python
for comic in com:
    comic_id.append(comic["id"])
```

```python
len(comic_id)
```

### Obtaining creator

```python
comic_creator = []
```

```python
for comic in com:
    comic_creator.append(", ".join([creator["name"] for creator in comic["creators"]["items"]]))
```

```python
len(comic_creator)
```

### Obtaining editors

```python
comic_editor = []

for comic in com:
    editor_name = "Not given"
    for creator in comic["creators"]["items"]:
        if creator["role"] == "editor":
            editor_name = creator["name"]
            break
    comic_editor.append(editor_name)
```

```python
len(comic_editor)
```

### Obtaining writers

```python
comic_writer = []


for comic in com:
    writer_names = []
    for writer in comic["creators"]["items"]:
        
        if writer["role"] == "writer":
            writer_names.append(writer["name"])
        
            
    comic_writer.append(writer_names)
```

```python
len(comic_writer)
```

### Obtaining format

```python
comic_format = []
```

```python
for comic in com:
    comic_format.append(comic["format"])
```

```python
len(comic_format)
```

### Obtaining on_sale_dates

```python
comic_dates = []
```

```python
for comic in com:
    
    comic_dates.append(comic["dates"])
```

```python
len(comic_dates)
```

```python
print(comic_dates[0])
```

```python
raw_dates_onsale = []
```

```python
for dates in comic_dates:
    for i in dates:
        if i["type"] == "onsaleDate":
            raw_dates_onsale.append(i["date"])   
```

```python
len(raw_dates_onsale)
```

```python
print(raw_dates_onsale[0])
```

```python
comic_dates_onsale = []

for date in raw_dates_onsale:
    onsale = date.split("T")[0]
    comic_dates_onsale.append(onsale)
```

### Obtaining prices

```python
raw_prices = []
comic_prices = []

for comic in com:
    raw_prices.append(comic["prices"])

for price in raw_prices:
    for i in price:
        if i["type"] == "printPrice":
            comic_prices.append(i["price"])
```

```python
len(comic_prices)
```

### Obtaining n_of_pages

```python
comic_pages = []

for comic in com:
    comic_pages.append(comic["pageCount"])
```

```python
len(comic_pages)
```

### Obtaining series

```python
comic_series = []

for comic in com:
    comic_series.append(comic["series"]["name"])
```

```python
len(comic_series)
```

## playing pandas

```python
marvel_df = pd.DataFrame({"title":comic_list,
                          "series":comic_series,
                          "on_sale_date":comic_dates_onsale,
                          "creators":comic_creator,
                          "creators_editor":comic_editor,
                          "creators_writer":comic_writer,
                          "description":comic_description, 
                          "characters":comic_characters, 
                          "format":comic_format,
                          "n_of_pages":comic_pages,
                          "price":comic_prices,
                         })
```
